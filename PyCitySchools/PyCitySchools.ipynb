```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
# data to Load
school_data_path = os.path.join('Resources','schools_complete.csv')
student_data_path = os.path.join('Resources','students_complete.csv')

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_path)
student_data = pd.read_csv(student_data_path)

# Combine the data into a single dataset
complete_school_data= pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
complete_school_data.head()
```

# District Summary

Create a high level snapshot (in table form) of the district's key metrics, including:

    Total Schools
    Total Students
    Total Budget
    Average Math Score
    Average Reading Score
    % Passing Math (The percentage of students that passed math.)
    % Passing Reading (The percentage of students that passed reading.)
    % Overall Passing (The percentage of students that passed math and reading.)


```python
#find total number of schools using the .unique method
total_schools = len(complete_school_data['school_name'].unique())
total_student = complete_school_data['student_name'].count()
total_budget = sum(complete_school_data['budget'].unique())
math_avg_score = complete_school_data['math_score'].mean()
reading_avg_score = complete_school_data['reading_score'].mean()
overall_passing_score = (math_avg_score + reading_avg_score)/2
math_pass_score = (complete_school_data[complete_school_data['math_score']>=70]['student_name'].count()/total_student)*100
reading_pass_score = (complete_school_data[complete_school_data['reading_score']>=70]['student_name'].count()/total_student)*100


#create new df to display the above data

district_df = {
    'Total Schools':total_schools,
    'Total Student':'{:,}'.format(total_student),
    'Total Budget':'${:,.2f}'.format(total_budget),
    'Avg Math Score':math_avg_score,
    'Avg Reading Score':reading_avg_score,
    '% Passing Math':math_pass_score,
    '% Passing Reading':reading_pass_score,
    '% Overall Passing Score':[overall_passing_score],  
}

district_sum_df = pd.DataFrame(district_df)
district_sum_df
```
