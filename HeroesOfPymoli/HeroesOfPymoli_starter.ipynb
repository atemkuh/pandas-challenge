```python
#
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head()
```

```python
#replace other / Non-diclosed with non_specific_gender
purchase_data_df.replace('Other / Non-Disclosed','non_specific_gender', inplace=True)
```

```python
# to verify change of 'Other / Non-Disclosed ' to 'non_specific_gender'
count = purchase_data_df["Gender"].value_counts()
count
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data_df["SN"].count()
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#caculate unique items, total purchase,revenue
unique_items = len(purchase_data_df["Item ID"].unique())
unique_items

sum_purchases = purchase_data_df["Purchase ID"].count()
sum_purchases

sum_revenue = purchase_data_df["Price"].sum()
sum_revenue

average_price1 = purchase_data_df["Price"].mean()
average_price1

average_price2 = sum_revenue/total_purchases
average_price2


# create a display to summarize  df
purchase_analysis_df = pd.DataFrame([{"Number of Unique Items": unique_items, "Average Price": average_price1,
                                      "Number of Purchases": total_purchases, "Total Revenue": total_revenue}])
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:,.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)
purchase_analysis_df

# create new column order for df summary
new_purchase_analysis_df = purchase_analysis_df[["Number of Unique Items","Average Price","Number of Purchases", "Total Revenue" ]]
new_purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# get unique values in a column by using the value count method
# then create a dataframe to hold results
gender_demo_df = pd.DataFrame(purchase_data_df["Gender"].value_counts())
gender_demo_df

percentage_of_players = (purchase_data_df["Gender"].value_counts()/total_players)*100
percentage_of_players

# Calculate and add results into DataFrame as a new column
gender_demo_df["Percentage of Players"] = percentage_of_players
gender_demo_df["Percentage of Players"] = gender_demo_df["Percentage of Players"].map("{:,.2f}%".format)
gender_demo_df

# re-order columns 
org_gender_demo_df = gender_demo_df[["Percentage of Players", "Gender"]]
org_gender_demo_df

# Rename the column "Gender" to "Total Players" using .rename(columns={})
fin_gender_demo_df = org_gender_demo_df.rename(columns={"Gender":"Total Players"})
fin_gender_demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Separate “GENDER”  into fields usinf Groupby syntax
gender_grouped_purchased_data_df = purchase_data_df.groupby(["Gender"])
gender_grouped_purchased_data_df["Purchase ID"].count().head(10)

# use gender to determine total purchase value 
total_purchase_value = gender_grouped_purchased_data_df["Price"].sum()
total_purchase_value.head()
dlr_total_purchase_value = total_purchase_value.map("${:,.2f}".format)
dlr_total_purchase_value.head()

# determine average purchase price by gender
avg_purchase_price = gender_grouped_purchased_data_df["Price"].mean()
avg_purchase_price.head()
dlr_avg_purchase_price = avg_purchase_price.map("${:,.2f}".format)
dlr_avg_purchase_price.head()

# find total by dividing total purchase by gender purchase 
# "avg_tpp" is same as or refered to as "avg_total_per_person"

avg_total_per_person_totals = total_purchase_value/gender_grouped_purchased_data_df["Purchase ID"].count()
dlr_avg_tpp_totals = avg_total_per_person_totals.map("${:,.2f}".format)
dlr_avg_tpp_totals.head()

# to get gender summary and display in new dataframe 
get_gender_purchased_data_df = pd.DataFrame(gender_grouped_purchased_data_df["Purchase ID"].count())
get_gender_purchased_data_df["Average Purchase Price"] = dlr_avg_purchase_price  
get_gender_purchased_data_df["Total Purchase Value"] = dlr_total_purchase_value 
get_gender_purchased_data_df["Avg Total Purchase per Person"] = dlr_avg_tpp_totals
get_gender_purchased_data_df

# Summary purchasing analysis DF grouped by gender
sum_gender_purchased_data_df = get_gender_purchased_data_df.rename(columns={"Purchase ID":"Purchase Count"})
sum_gender_purchased_data_df
```
